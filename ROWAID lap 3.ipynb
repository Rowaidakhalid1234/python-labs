```python
color = input("Enter color: ")

if color != "":
    match color:
        case "red":
            print("stop")
        case "yellow":
            print("Get Ready")
        case "green":
            print("go")
        case _:
            print("Unknown color")
else:
    print("Empty input")





```

```python
score = int(input("Enter your score: "))

if score >= 50:
    print("Passed")
else:
    print("Failed")

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Pass")
    case "F":
        print("Fail")
```

```python
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = input("Choose option: ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

match choice:
    case "1":
        print(num1 + num2)
    case "2":
        print(num1 - num2)
    case "3":
        print(num1 * num2)
    case "4":
        print(num1 / num2)
    case _:
        print("Invalid choice")
```
