```python
# Question 1
import math

def circle_area(radius):
    return math.pi * radius * radius

circumference = lambda r: 2 * math.pi * r

radius = float(input("Enter radius of the circle: "))

area = circle_area(radius)
circ = circumference(radius)

print("Area =", round(area, 2))
print("Circumference =", round(circ, 2))
```

```python
# Question 2
import re

def sum_numbers(nums):
    return sum(nums)

text = input("Enter a string: ")

numbers = re.findall(r'\d+', text)
numbers = list(map(int, numbers))

total = sum_numbers(numbers)
double = lambda x: x * 2

print("Numbers found:", numbers)
print("Sum:", total)
print("Doubled value:", double(total))
```

```python
# Question 3
import math

nums_input = input("Enter numbers separated by commas: ")
nums_list = list(map(float, nums_input.split(",")))

def square_list(lst):
    return list(map(lambda x: x**2, lst))

squares = square_list(nums_list)

max_square = max(squares)
sqrt_sum = math.sqrt(sum(squares))

print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sum)
```
