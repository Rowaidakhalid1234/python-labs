```python
print("Read 10 Numbers :")
counter = 0
sum = 0

while True:
    number = float(input("Read number"))

    if number < 0:
        print("can not Read Negative Number :")
        continue

    if number == 0:
        break

    if number % 2 == 0:
        sum += number

    counter += 1
    if counter == 10:
        break

print("Sum of Even Numbers =", sum)
```

```python
print("Read Number to Generate Multiplication Tabel :")
number = int(input("Read number"))

for x in range(1, number+1, 1):
    if x == 3:
        continue

    product = x * 12
    if product >= 50:
        break

    print(x, "x", "12", "=", (x*12))
```

```python
counter = 0

number = int(input("Enter Number :"))

for x in range(1, number + 1, 1):

    if number > 50:
        print("Stop Counting ")
        break

    if x % 4 == 0:
        counter += 1

    if x % 8 == 0:
        continue

print("Number Divided by 4 =", counter)
```
