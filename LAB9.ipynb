```python
grades = np.array([[87, 96, 70],
                   [100, 87, 90],
                   [94, 77, 90],
                   [100, 81, 82]])

print("sum =", np.sum(grades))
print("min =", np.min(grades))
print("max =", np.max(grades))
print("mean =", np.mean(grades))
print("std =", np.std(grades))
print("var =", np.var(grades))

print("mean per column =", np.mean(grades, axis=0))
print("mean per row =", np.mean(grades, axis=1))
print("max per row =", np.max(grades, axis=1))
```

```python
import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60]])

B = np.array([1, 2, 3])

# Add B to A
add_result = A + B
print("A + B =\n", add_result)

# Multiply A by 2
mul_result = A * 2
print("A * 2 =\n", mul_result)

# Subtract 5 from A
sub_result = A - 5
print("A - 5 =\n", sub_result)
```

```python
grades = np.array([[70, 85, 90],
                   [88, 76, 95],
                   [60, 80, 72]])

# Index of largest value in whole array
max_index = np.argmax(grades)
print("Index of largest value (flattened):", max_index)

# Index of highest grade in each row
row_max_index = np.argmax(grades, axis=1)
print("Index of max in each row:", row_max_index)

# Index of lowest grade in each column
col_min_index = np.argmin(grades, axis=0)
print("Index of min in each column:", col_min_index)
```

```python
import numpy as np

x = np.array([2, 0, 1, 5, 4, 1, 9])
A = np.array([[3, 8, 1],
              [6, 2, 9]])

# Part 1: argsort
indices = np.argsort(x)
sorted_x = x[indices]

print(indices)
print(sorted_x)

# Part 2: min location
min_index = np.unravel_index(A.argmin(), A.shape)
print(min_index)

# Part 3: where
indices_where = np.where(x > 3)
print(indices_where)

# Part 4: split
parts = np.split(x, [2, 5])
for part in parts:
    print(part)

# Part 5: expand and squeeze
expanded = np.expand_dims(x, axis=0)
print(expanded)

squeezed = np.squeeze(expanded)
print(squeezed)
```
