```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    
    result = num1 / num2

except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

except ValueError:
    print("Error: Please enter valid numbers (not text).")

else:
    print("Result =", result)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name      # Public property
        self._age = age       # Property variable (private style)

# Create object
p1 = Person("Rowaida", 20)

# Update age
p1._age = 22

# Display updated details
print("Name:", p1.name)
print("Updated Age:", p1._age)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary += bonus

    def display(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)


# Create object
m1 = Manager("Rowaida", 5000)

# Add bonus
m1.add_bonus(1000)

# Display information
m1.display()
```
